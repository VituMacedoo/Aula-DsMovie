```python
import yfinance as yf
```

```python
teste_merge = yf.download ('MGLU3.SA', start='2020-01-01', end= '2020-07-01')
```

```python
weg_teste= yf.download('WEGE3.SA', start = '2020-01-01', end = '2020-10-01')
```

```python
weg_teste
```

```python
weg_teste.rename(columns = {'Adj Close': 'WEGE3'},inplace= True)
```

```python
teste_merge.rename(columns = {'Adj Close': 'MGLU3'},inplace= True)
```

```python
teste_merge
```

```python
#merge = fusao = join
novo_df = pd.merge(weg_teste, teste_merge, how = 'inner', on= 'Date')
```

```python
novo_df.head(10)
```

```python
cotacoes = novo_df[['WEGE3', 'MGLU3']]
```

```python
cotacoes.head()
```

```python
cotacoes.plot()
```
